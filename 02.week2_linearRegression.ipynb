```python
import tensorflow as tf
import numpy as np
```

```python
x_train = [1, 2, 3, 4, 5]
y_train = [6, 7, 8, 9, 10]

# shape 1 (rank 1)
W = tf.Variable(tf.random.normal([1]), name='weight')
b = tf.Variable(tf.random.normal([1]), name='bias')
# H
hypothesis = x_train * W + b
```

```python
# cost(W, b)
cost = tf.reduce_mean(tf.square(hypothesis - y_train))
print(cost)
```

```python
#optimizaer = tf.train.GradientDescentOptimizer(learning_late=0.01)
sgd = tf.keras.optimizers.SGD(learning_rate=0.005)

model = tf.keras.models.Sequential() #순서대로 만든 모델
model.add(tf.keras.layers.Dense(1, input_dim = 1)) # 레이어 추가, 
model.compile(loss='mean_squared_error', optimizer = sgd)

model.fit(x_train, y_train, epochs=500, verbose=2) #verbose 0 로그x 1 로그o 2 간결로그o

print(model.predict([1]))
print(W)
print(b)
```
